Armstrong number

```python
num = int(input("enter the number"))
temp = num
total = 0
l=len(str(num))
while (num > 0):
  digit = num % 10
  total += digit ** l
  num = num //10
if (total == temp):
  print("armstrong")
else:
  print("not armstrong")
```

### Student Report Card Program

```python
# 1. Get student name and age
student_name = input("Enter student name: ")
student_age = int(input("Enter student age: "))
```

```python
# 2. Get number of subjects and subject names
num_subjects = int(input("Enter the number of subjects: "))
subjects = set()
for i in range(num_subjects):
  subject_name = input(f"Enter name for subject {i+1}: ")
  subjects.add(subject_name)

print(f"Subjects entered: {subjects}")
```

```python
# 3. Get subject-wise marks
subject_marks = {}
for subject in subjects:
  marks = float(input(f"Enter marks for {subject}: "))
  subject_marks[subject] = marks

print(f"Subject-wise marks: {subject_marks}")
```

```python
# 4. Store data in specified data structures

# Name in a list
student_names_list = [student_name]

# Subjects in a set (already handled in previous step)
# subjects_set = subjects

# Student info in a tuple
student_info_tuple = (student_name, student_age)

# Marks in a dictionary (already handled in previous step)
# subject_marks_dict = subject_marks

print(f"\nStored Data:")
print(f"Student Name (List): {student_names_list}")
print(f"Student Info (Tuple): {student_info_tuple}")
print(f"Subjects (Set): {subjects}")
print(f"Subject Marks (Dictionary): {subject_marks}")
```

```python
# 5. Print a report card

print("\n--- Student Report Card ---")
print(f"Name: {student_info_tuple[0]}")
print(f"Age: {student_info_tuple[1]}")
print("Subjects & Marks:")

total_marks = 0
for subject, marks in subject_marks.items():
  print(f"  - {subject}: {marks}")
  total_marks += marks

# Calculate average marks
num_subjects_recorded = len(subject_marks)
if num_subjects_recorded > 0:
  average_marks = total_marks / num_subjects_recorded
else:
  average_marks = 0 # Handle case with no subjects/marks

print(f"Total Marks: {total_marks}")
print(f"Average Marks: {average_marks:.2f}")
print("-------------------------")
```
